# Step 1: Import Required Libraries


```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
print("Libraries successfully import ho gayi hain!")
```

# Step 2: Load the IMDB Dataset


```python
vocab_size = 10000  # Sirf top 10,000 mashhoor lafz use karenge
max_len = 100       # Har review ko 100 lafzon tak mahdood karenge

print("Data load ho raha hai...")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)
print(f"Training reviews: {len(X_train)}")
print(f"Testing reviews: {len(X_test)}")
```

# Step 3: Pad Sequences to Uniform Length


```python
print("Padding shuru...")
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
print("Sab reviews ka size ab 100 words ho chuka hai!")
```

# Step 4: Build the LSTM Model Architecture


```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_len),
    LSTM(units=32),
    Dense(units=1, activation='sigmoid') # 1 neuron positive (1) ya negative (0) batane ke liye
])

print("Model summary neechay majood hai:")
model.summary()
```

# Step 5: Compile and Train the Model


```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Model ki training shuru ho rahi hai, thoda intezar karein...")
history = model.fit(X_train, y_train, epochs=2, batch_size=64, validation_data=(X_test, y_test))
```

# Step 6: Evaluate Model Performance and Accuracy


```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n--- Assignment Final Result ---")
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
