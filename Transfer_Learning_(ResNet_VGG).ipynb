# Step 1: Import Required Libraries


```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.utils import to_categorical
print("Libraries successfully import ho gayi hain!")
```

# Step 2: Load and Preprocess CIFAR-10 Dataset


```python
print("Data load ho raha hai...")
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Images ko normalise karna (0 to 1 range)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Labels ko One-Hot Encode karna (10 classes hain)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
print("Data ready hai!")
```

# Step 3: Load Pre-trained VGG16 Base Model


```python
print("Pre-trained VGG16 model load ho raha hai...")
# include_top=False ka matlab hai hum classification layer khud lagayenge
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Purane weights ko freeze kar rahe hain taake training kharab na ho
base_model.trainable = False
print("VGG16 Base successfully freeze ho gaya!")
```

# Step 4: Build Custom Transfer Learning Model Architecture


```python
model = Sequential([
    base_model,                     # Pre-trained VGG16 base
    GlobalAveragePooling2D(),       # Features compress karne ke liye
    Dense(128, activation='relu'),   # Naya hidden layer
    Dense(10, activation='softmax') # 10 classes ke liye final output
])

print("Model summary:")
model.summary()
```

# Step 5: Compile and Train the Model


```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("Model ki training shuru ho rahi hai (Sirf naye layers train honge)...")
# Epochs=2 rakha hai taake jaldi complete ho jaye
model.fit(X_train, y_train, epochs=2, batch_size=64, validation_data=(X_test, y_test))
```

# Step 6: Evaluate Model Performance and Accuracy


```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n--- Assignment Final Result ---")
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
